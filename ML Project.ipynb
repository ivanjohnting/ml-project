```python
import re
from collections import Counter
from collections import namedtuple

train = open('train', encoding = "utf8") #open the training file

train_xlist=[] #create list to store the emission values
train_ylist=[] #create list to store the transistion states
train_xylist=[] #note the instances from y to x

for lines in train:
	line=lines.split() #create array for elements
	#print (line)
	if line!=[]:
		train_xlist.append(line[0])
		train_ylist.append(line[1])
		train_xylist.append(line)
train_xcount = Counter(train_xlist) #count the number of times an element in the list appears (dictionary)
train_ycount = Counter(train_ylist)
train_xycount = Counter(tuple(i) for i in train_xylist) #since this is a nested list
# print (train_xcount)
print (train_ycount)
#print(train_xycount)

# emissiondict = {} #create the instance of the dictionary
# for key in train_xycount: #for each instance of y to x
# 	emission = train_xycount[key]/train_ycount[key[1]] #calculate the emission value
# 	emissiondict[key] = emission #store the probability in the dictionary
# #print (emissiondict)

#section to change the emission parameters to accomodate for unseen words
emissiondict = {} #create the instance of the dictionary
for key in train_xycount: #for each instance of y to x
	emission = train_xycount[key]/(train_ycount[key[1]]+1) #calculate the emission value
	emissiondict[key] = emission #store the probability in the dictionary

for key in train_ycount: #for alll the states, get the "unknown" probability
    emissiondict[("unknown",str(key))]=1/(train_ycount[key]+1)
    
#print (emissiondict)


test = open('dev.in', encoding = "utf8") #open the unlabelled data
test_xlist=[] #create empty list to store all the emission values

for lines in test:
    line2=lines.split()
    if line2!=[]:
        test_xlist.append(line2[0])
        
for x in test_xlist:
    if x not in train_xcount:
        for n,i in enumerate(test_xlist):
            if i==x:
                test_xlist[n]="unknown"



def emission(train_xycount,train_ycount):
    
    emissiondict={}
    for key in train_xycount:
        emission=train_xycount[key]/train_ycount[key[1]]
        emissiondict[key]=emission
    
    return emissiondict

#emission(train_xycount,train_ycount)        

    
#print (train_ycount)
```

```python
#f = open("C:\\Users\\Alex\\Documents\\dev.p2.out", 'w')
ylisted=[]
for x in test_xlist:
    label=[]
    scores=[]
    
    for key in emissiondict:
        if key[0]==x:
            label.append(key) #key
            scores.append(emissiondict[key]) #value
    index=scores.index(max(scores))
    #f.write(str(label[index]))
    ylisted.append(label[index][1])
    #print(label[index]) #print the best label

print (ylisted) #where ylisted is the list of predicted sentiments

#f.close



    
```
